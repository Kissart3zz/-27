```python
def max_profit(prices):
    min_price = prices[0]
    buy_day = 0

    max_profit = 0
    sell_day = 0

    for i in range(1, len(prices)):
        profit = prices[i] - min_price

        if profit > max_profit:
            max_profit = profit
            sell_day = i

        if prices[i] < min_price:
            min_price = prices[i]
            buy_day = i

    if max_profit > 0:
        print("Покупать в день", buy_day + 1)
        print("Продавать в день", sell_day + 1)
        print("Прибыль:", max_profit)
    else:
        print("Прибыль невозможна")


prices = [9, 1, 5]

max_profit(prices)
```
