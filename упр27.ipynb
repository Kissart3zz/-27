```python
def find_peak(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        mid = (left + right) // 2

        if arr[mid] < arr[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return left


A = [1, 3, 7, 12, 9, 5, 2]

p = find_peak(A)

print(p)
print(A[p])
```
